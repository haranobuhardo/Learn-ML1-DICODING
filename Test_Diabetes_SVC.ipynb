# 1. Data Processing

```python
df = pd.read_csv('datasets/diabetes.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
X = df[df.columns[:8]]
y = df['Outcome']
print(X.head())
print(y[:5])
```

# 2. Data Preparing

```python
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
#X = pd.DataFrame(X, columns=df.columns[:8])
X
```

#3. Data Splitting

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.33, random_state=42)
```

# 4. Data Modelling

```python
clf = SVC()
clf.fit(X_train, y_train)
```

# 5. Data Evaluation

```python
clf.score(X_test, y_test)
```
