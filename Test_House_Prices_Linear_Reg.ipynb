```python
import numpy as np
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
```

# 1. Data Processing

```python
# buat data jumlah kamar
bedrooms = np.array([1,1,2,2,3,4,4,5,5,5])

# buat data harga rumah (asumsi dalam dollar)
house_price = np.array([15000, 18000, 27000, 34000, 50000, 68000, 65000, 81000,85000, 90000])

bedrooms.shape == house_price.shape
```

```python
plt.scatter(bedrooms, house_price)
plt.xlabel('Bedrooms Number')
plt.ylabel('House Prices (in $)')
plt.show()
```

# 2. Data Modelling

```python
bedrooms = bedrooms.reshape(-1, 1)
linreg = LinearRegression()
linreg.fit(bedrooms, house_price)
```

# 3. Data Visualization

```python
plt.scatter(bedrooms, house_price)
plt.plot(bedrooms, linreg.predict(bedrooms), color='black', label='linreg')
plt.xlabel('Bedrooms Number')
plt.ylabel('House Prices (in $)')
plt.legend()
plt.show()
```

# 4. Model Evaluation

```python
from sklearn import metrics
import numpy as np

y_pred = linreg.predict(bedrooms)

print('R-Squared value of the model', metrics.r2_score(house_price, y_pred))
print('R-Squared value of the model', linreg.score(bedrooms, house_price))
print('RMSE', np.sqrt(metrics.mean_squared_error(house_price, y_pred)))
```
