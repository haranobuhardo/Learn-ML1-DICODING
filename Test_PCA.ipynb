```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.decomposition import PCA
```

# 1. Data Processing

```python
iris = datasets.load_iris()
atribut = iris.data
label = iris.target
print(atribut[:5])
print(label[:5])
```

# 2. Data Splitting

```python
X_train, X_test, y_train, y_test = train_test_split(atribut, label, test_size=0.2, random_state=1)
```

# 3. Data Modelling
with default attributes DecisionTreeClassifier()

```python
decision_tree = tree.DecisionTreeClassifier()
first_model = decision_tree.fit(X_train, y_train)
```

## 3.1 Data Evaluation

```python
first_model.score(X_test, y_test)
```

# 4. Dimensionality Reduction

List variance for all attributes

```python
pca = PCA(n_components=4) # PCA with 4 principal components or attributes

pca.fit_transform(X_train)

pca.explained_variance_ratio_
```

From the result above, we are going to use only 2 principal components or attributes

Create attributes data with only 2 principal components or attributes

```python
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.fit_transform(X_test)
```

## 4.1 PCA Data Evaluation

```python
second_model = decision_tree.fit(X_train_pca, y_train)
second_model.score(X_test_pca, y_test)
```
