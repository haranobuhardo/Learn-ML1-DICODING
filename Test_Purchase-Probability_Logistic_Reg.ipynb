# 1. Data Processing

```python
df = pd.read_csv('datasets/Social_Network_Ads.csv')
df
```

```python
print(df.info())
print(df.describe())
```

# 2. Data Preparation
Tranforming categorial data to numerical data

```python
  df = pd.get_dummies(df)
  df
```

Dividing between attributs (X) and label (y)

```python
X = df[['Age', 'EstimatedSalary', 'Gender_Female', 'Gender_Male']]
y = df['Purchased']
```

**Standardization** using StandardScaler()

```python
scaler = StandardScaler()
scaler.fit(X)
scaled_data = scaler.transform(X)
scaled_data = pd.DataFrame(scaled_data, columns=X.columns)
scaled_data
```

# 3. Data Train and Test Splitting

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_data, y, test_size=0.2, random_state=1)
```

#4. Data Modelling
with LogisticRegression()

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

# 5. Data Evaluation

```python
print('The accuracy of the model is', model.score(X_test, y_test))
```
