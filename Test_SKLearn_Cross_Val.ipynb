```python
import sklearn
from sklearn import datasets
from sklearn import tree
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data
y = iris.target
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
scores = cross_val_score(clf, X, y, cv=5)
scores
```

```python
plt.plot([x+1 for x in range(len(scores))], scores)
# or you can go with plt.plot(scores)
plt.xlabel('CV Value')
plt.ylabel('Accuracy')
plt.title('Testing Cross Validation Accuracy Score')
plt.style.use('fast')
plt.show()
```
