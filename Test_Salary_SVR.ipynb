# 1. Data Processing

```python
df = pd.read_csv('datasets/Salary_Data.csv')
df
```

```python
X = df['YearsExperience']
y = df['Salary']

print(X[:5])
print(y[:5])

X = X.to_numpy().reshape(-1, 1)
X
```

# 2. Data Modelling

```python
model = SVR(C=1000, gamma = 0.05, kernel='rbf')

model.fit(X, y)
```

# 3. Data Visualization

```python
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
